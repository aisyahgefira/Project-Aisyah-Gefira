```python
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 11223 KJ',
                   'AE 44677 GH', 'AE 67269 AA']

#buat list plat mobil yang hanya berisi angka saja
plat_baru = []

for i in list_plat_mobil:
    split_name = i.split()
    new_name = split_name[1]
    plat_baru.append(new_name)

#konversi list ke integer
plat_baru = [int(x) for x in plat_baru]

#menentukan plat yang ganjil dan genap
ganjil=[]
genap=[]
x=0

while x<len(plat_baru):
    if plat_baru[x]%2==0:
        genap.append(plat_baru[x])
    else:
        ganjil.append(plat_baru[x])
    x+=1
jumlah_genap=len(genap)
jumlah_ganjil=len(ganjil)

print(f"jumlah plat genap: {jumlah_genap}")
print(f"jumlah plat ganjil: {jumlah_ganjil}")
```

```python
count = 0
n1 = 2   #dimulai dari 2 dan 3
n2 = 3   #dimulai dari 2 dan 3

fibonacci = []
while count < 12:          #deret fibonacci 12 angka
    fibonacci.append(n1)
    last_num = n1 + n2
    n1 = n2
    n2 = last_num
    count += 1

print(fibonacci)
```

```python
data = input("Masukkan suatu kata: ")
proses = input("a = huruf pertama | b = huruf terakhir | c = jumlah huruf | d = jumlah huruf voca; || Ingin cari tahu apa (a,b,c,d)?:  ")

def string_processor(data, proses):
    if proses == 'a':
        first_char = data[0]
        return first_char
    elif proses == 'b':
        last_char = data[len(data)-1]
        return last_char
    elif proses == 'c':
        num_chars = len(data)
        return num_chars
    else:
        vocal = {'a':0, 'i':0, 'u':0, 'e':0, 'o':0}
        num_vowels = 0
        for i in data:
            if i in ['a', 'i', 'u', 'e', 'o']:
                vocal[i] += 1
                num_vowels += 1
        print(num_vowels)

string_processor(data,proses)
```

```python
#rename kolom
penjualan = penjualan.rename(columns={'Row ID':'row_id',
                                     'Order ID':'order_id',
                                     'Order Date': 'order_date',
                                     'Ship Date': 'ship_date',
                                     'Ship Mode': 'ship_mode',
                                     'Customer ID': 'customer_id',
                                     'Customer Name': 'customer_name',
                                     'Postal Code': 'postal_code',
                                     'Product ID': 'product_id',
                                     'Product Name': 'product_name'})
```

```python
penjualan
```

```python
default = penjualan['order_date'][0:4] == "2017"
def selling(Region, default):
    selected_region_year = penjualan[(penjualan.order_date == default)
                                    & (penjualan.Region == Region)].reset_index()
    total_sales = selected_region_year.Sales.sum()
    total_profit = selected_region_year.Profit.sum()
    print('Total penjualan pada', default, 'untuk daerah', Region, 'adalah', total_sales)
    print('Total keuntungan pada', default, 'untuk daerah', Region, 'adalah', total_profit)
selling('West','2017')
selling('South', '2016')
```

```python
def segment(x):
    if x['Sales'] > 1500000:
        segment = 'high'
    elif x['Sales'] > 765000 and x['Sales'] <= 1500000:
        segment = 'mid'
    else:
        segment = 'low'
    return segment
penjualan['performance_sales'] = penjualan.apply(lambda x: segment(x), axis = 1)
penjualan
```

```python
default = penjualan['order_date'][0:4] == "2017"
def selling(performance, default):
    selected_region_year = penjualan[(penjualan.order_date == default)
                                    & (penjualan.performance_sales == performance)].reset_index()
    total_cust = selected_region_year.customer_id.count()
    print('Total customer pada', default, 'untuk segment', performance, 'adalah', total_cust)
selling('low','2017')
selling('mid', '2017')
selling('high', '2017')
```
